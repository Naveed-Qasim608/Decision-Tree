BUILDING A DECISION TREE



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

# Load CSV dataset
data = pd.read_csv('calorie_efficiency_dataset.csv')

# Convert into NumPy array
data = data.values

# Separate features and labels
X = data[:, :-1]
y = data[:, -1]

# Separate classes
class_0 = X[y == 0]
class_1 = X[y == 1]

# Plot dataset
plt.figure(figsize=(8, 6))

plt.scatter(class_0[:, 0], class_0[:, 1],
            s=75,
            color='black',
            marker='x',
            label='Class 0')

plt.scatter(class_1[:, 0], class_1[:, 1],
            s=75,
            facecolors='white',
            edgecolors='black',
            marker='o',
            label='Class 1')

plt.title('Calorie Efficiency Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)
plt.legend()

plt.show()

# Split training and testing data
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=5
)

# Create Decision Tree classifier
classifier = DecisionTreeClassifier(
    random_state=0,
    max_depth=4
)

# Train the model
classifier.fit(X_train, y_train)

# Predict test data
y_test_pred = classifier.predict(X_test)

# Display classification report
print("\nClassification Report:\n")

print(classification_report(y_test, y_test_pred))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder


# 1. Load CSV Data

input_file = 'calorie_efficiency_dataset.csv'
data = pd.read_csv(input_file)

print("First 5 rows:\n")
print(data.head())


# 2. Separate Features and Label

# Last column is target
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values


# 3. Convert String Labels to Numbers

le = LabelEncoder()
y = le.fit_transform(y)

print("\nClasses:", le.classes_)


# 4. Separate Classes (for plotting)

class_0 = X[y == 0]
class_1 = X[y == 1]


# 5. Visualize Data

plt.figure()

plt.scatter(class_0[:, 0], class_0[:, 1],
            s=75, color='black', marker='x', label='Class 0')

plt.scatter(class_1[:, 0], class_1[:, 1],
            s=75, facecolors='white', edgecolors='black',
            marker='o', label='Class 1')

plt.title("Input Data (First 2 Features)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()


# 6. Train-Test Split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=5
)


# 7. Build Decision Tree Model

params = {'random_state': 0, 'max_depth': 4}
classifier = DecisionTreeClassifier(**params)

classifier.fit(X_train, y_train)


# 8. Predictions

y_test_pred = classifier.predict(X_test)


# 9. Evaluation

print("\nClassification Report:\n")
print(classification_report(y_test, y_test_pred))
```
